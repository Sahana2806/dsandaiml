```python
def mean(array):
  n=len(array)
  return np.sum(array)/n
```

```python
mean(np.array([1,2,3,4,5,6]))
```

```python
def meadian(array):
  n=len(array)
  return np.sum(array)/n
```

```python
meadian(np.array([1,2,3,4,5]))
```

```python
def mode(array):
  n=len(array)
  return np.sum(array)/n
```

```python
mode(np.array([1,2,3,4,5]))
```

```python
def variance(array):
  n=len(array)
  m=mean(array)
  return np.sum((array - m)**2)/n
```

```python
variance(np.array([1,2,3,4,5,6]))
```
